# Adım 2: Kafka ile Streaming Veri Üretimi

Bu aşamada, seçilen veri setinden gerçek zamanlı veri akışı simülasyonu yapan bir Python Kafka Producer geliştirilmiştir. Producer, CSV dosyasından satırları okuyarak belirli aralıklarla Kafka topic'ine JSON formatında mesaj göndermektedir.

## 2.1. Producer Mantığı ve Gereksinimler

Gereksinimlere uygun olarak geliştirilen Producer şu özelliklere sahiptir:
- **JSON Formatı:** Mesajlar JSON olarak serileştirilir.
- **Zorunlu Alanlar:** Her mesajda `timestamp`, `kullanici_id`, `olay_tipi` ve `ilgili_id` alanları bulunur.
- **Ayarlanabilir Hız:** Saniyede gönderilecek mesaj sayısı `MSG_PER_SECOND` ortam değişkeni ile kontrol edilebilir.
- **Loglama:** Gönderilen toplam mesaj sayısı loglar üzerinden takip edilebilir.
