## 6.1. Eğitilen Modeller

Aşağıdaki 5 farklı algoritma kullanılarak modelleme yapılmıştır:
1.  **Linear Regression:** Temel doğrusal regresyon modeli.
2.  **Decision Tree Regressor:** Karar ağacı tabanlı model.
3.  **Random Forest Regressor:** Topluluk (Ensemble) öğrenme modeli.
4.  **Gradient Boosted Trees (GBT) Regressor:** Güçlü boosting algoritması.
5.  **Generalized Linear Regression (GLR):** Genelleştirilmiş doğrusal model.

## 6.2. Değerlendirme Metrikleri

Modeller şu metrikler üzerinden karşılaştırılmıştır:
- **RMSE (Root Mean Squared Error):** Hataların karelerinin ortalamasının karekökü.
- **MAE (Mean Absolute Error):** Ortalama mutlak hata.
- **R² Score:** Modelin veriyi açıklama oranı.
