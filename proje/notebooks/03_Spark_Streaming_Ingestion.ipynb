# Adım 3: Spark Structured Streaming ile Veri Okuma

Kafka'dan gelen streaming veriyi Spark Structured Streaming ile okuyup işleyen ETL pipeline'ı bu aşamada kurulmuştur. Veriler Bronze, Silver ve Gold katmanları (Medallion Architecture) prensibine göre Delta Lake tablolarına yazılmaktadır.

## 3.1. Veri Katmanları (Medallion Architecture)

1.  **Bronze (Ham Veri):** Kafka'dan gelen mesajlar hiçbir değişikliğe uğramadan ham haliyle saklanır.
2.  **Silver (Temizlenmiş Veri):** JSON verisi parse edilir, şema uygulanır, null değerler filtrelenir ve duplike kayıtlar (watermark kullanılarak) temizlenir.
3.  **Gold (Analitik Veri):** İş birimi ihtiyaçlarına göre agregasyonlar (örneğin dakikalık işlem sayıları) yapılır.

## 3.3. Delta Lake Avantajları

Bu projede depolama formatı olarak Delta Lake tercih edilmiştir çünkü:
- **ACID Uyumluluğu:** Streaming yazma işlemleri sırasında veri bütünlüğü korunur.
