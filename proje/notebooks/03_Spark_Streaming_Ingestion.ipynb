## 3.1. Uygulanan İşlemler
1. **Kafka Entegrasyonu:** JSON formatındaki mesajlar Kafka üzerinden dinlenmiştir.
2. **Şema Tanımlama:** Veri tipi uyumluluğu için şema (`schema`) tanımlanmıştır.
3. **Bronze Katmanı:** Ham veriler (raw payload) doğrudan kaydedilmiştir.
4. **Silver Katmanı (Veri Temizleme):** 
   - Null değerler filtrelenmiştir.
   - Tekrar eden kayıtlar (duplicates) kaldırılmıştır.
   - Hatalı Quantity değerleri (negatifler) temizlenmiştir.
5. **Gold Katmanı:** İş birimi için gerekli olan saatlik gelir ve işlem sayıları özetlenmiştir.
