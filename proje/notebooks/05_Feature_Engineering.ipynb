# Adım 5: Özellik Mühendisliği (Feature Engineering)

Makine öğrenmesi modellerine beslenecek anlamlı özelliklerin üretildiği aşamadır. Bu projede veri sızıntısını (data leakage) önlemek amacıyla **Dinamik Zaman Bölmesi (Time-based Split)** yaklaşımı kullanılmıştır.

## 5.1. Özelliklerin Mantığı ve Seçimi

Modelde kullanılan temel özellikler ve iş mantıkları şunlardır:

1.  **Total Interactions:** Kullanıcının toplam işlem sayısı (Genel aktivite düzeyi).
2.  **Unique Items Viewed:** Kullanıcının ilgilendiği farklı ürün sayısı (Ürün çeşitliliği).
3.  **Active Days Count:** Kullanıcının aktif olduğu gün sayısı (Bağlılık/Loyalty).
4.  **Recency Days:** Kullanıcının son işleminden bu yana geçen gün sayısı (Tazelik).
5.  **Avg Daily Activity:** Günlük ortalama işlem sayısı.

**Hedef Değişken (Target):** Gelecek dönemdeki toplam harcama tutarı (`total_spending`).

## 5.2. Veri Sızıntısını Önleme Stratejisi

Veri seti zamansal olarak %80 (Geçmiş) ve %20 (Gelecek) olarak bölünmüştür:
- Özellikler sadece **ilk %80'lik** zaman diliminden hesaplanır.
- Hedef değişken ise sadece **son %20'lik** zaman diliminden hesaplanır.
Bu sayede modelin eğitim sırasında "geleceği görmesi" engellenmiş olur.
